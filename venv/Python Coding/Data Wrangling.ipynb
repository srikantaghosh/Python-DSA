```python
### replace each state names with obbreviation in dictionary

for state_name in us_state_to_abbrev:
    df_sample['state'].replace(to_replace=state_name, value=us_state_to_abbrev[state_name], inplace=True)
    
```

```python
### Rerun to confirm if changes took place
state_unique_values= df_sample['state'].unique()
state_unique_values.sort()
state_unique_values
```

###  Merging

```python
## we will merge the US abb dictionary to current dataframe
## Convert us_state_to_abbrev dictionary to dataframe

states_df = pd.DataFrame(list(us_state_to_abbrev.items()), columns=['state_name','state'])
states_df.head()
```

```python
## Merge the original dataframe with states dataframe

df_sample = pd.merge(df_sample,states_df, on='state', how='left')

df_sample.head()
```

```python
### rename 'state' column to 'state abbreviation'

df_sample.rename(columns={'state':'state_abbreviation'},inplace=True)

df_sample.head(3)
```

```python
## If we want to arrange the columns, like state_abbreviation and state_name side by side,we will use
## pop() and insert()

state_name_col = df_sample.pop('state_name')
df_sample.insert(3, 'state_name', state_name_col)
df_sample.head()
```

```python
## Clean the zip code column

zip_unique_values= df_sample['zip'].unique()
zip_unique_values.sort()
zip_unique_values
```

```python
## truncate all zip codes with more than five-digits down to five

df_sample['zip'] = df_sample['zip'].str[:5]
df_sample.head()
```

### Boolean indexing

```python
## create a boolean index to check for zip codes with a lenght of five digits
zip_code_boolean = df_sample['zip'].str.len()== 5


## display results of boolean index
zip_code_boolean
```

```python
## drop records that have a zip code length not equal to five

df_sample =  df_sample[zip_code_boolean]
```

```python
## get unique values for zip
zip_unique_values= df_sample['zip'].unique()
zip_unique_values.sort()
zip_unique_values
```
